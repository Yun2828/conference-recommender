### The purpose of this file is to add 20 abstracts each from INFORMS and IISE to the gemini-generated pre-existing test prompts list, which will be used to evaluate the model's retrieval and reranking.

#### This script takes 20 abstracts at random from the IISE and INFORMS datasets and appends them to the testing_queries file.

```python
import pandas as pd
```

```python
iise_dataset_filepath = './Datasets/IISE 2024/IISE2024_Processed.csv'
informs_dataset_filepath = r'.\Datasets\INFORMS 2024\INFORMS2024_processed.csv'
iise_queries_filepath = './Datasets/IISE 2024/Queries/testing_queries.xlsx'
informs_queries_filepath = r'./Datasets/INFORMS 2024/Queries/testing_queries.xlsx'

iise_df = pd.read_csv(iise_dataset_filepath)
informs_df = pd.read_csv(informs_dataset_filepath)

iise_queries = pd.read_excel(iise_queries_filepath, header=None)
informs_queries = pd.read_excel(informs_queries_filepath, header=None)
iise_queries.rename(columns={iise_queries.columns[0]: "Prompt"}, inplace=True)
informs_queries.rename(columns={informs_queries.columns[0]: "Prompt"}, inplace=True)

display(iise_queries.head(5))
display(informs_queries.head(5))

display(iise_df.head(5))
display(informs_df.head(5))
```

```python
def sample_and_append_rows(
    source_df: pd.DataFrame,
    source_column_name: str,
    n: int,
    target_df: pd.DataFrame,
    target_column_name: str,
    replace: bool = False
) -> pd.DataFrame:
    """
    Samples n random values from a source DataFrame column and appends them
    as new rows to a target DataFrame under a specified column name.

    Args:
        source_df: The pandas DataFrame to sample from.
        source_column_name: The name of the column in source_df to sample values from.
        n: The number of random values to sample.
        target_df: The pandas DataFrame to append the sampled values to.
        target_column_name: The name of the column in target_df where the
                             sampled values will be added. If this column doesn't
                             exist in target_df, it will be created, with NaN
                             values for the original rows.
        replace: Whether to sample with replacement (allow picking the same
                 source value multiple times). Defaults to False.

    Returns:
        A new pandas DataFrame containing the original target_df rows plus
        n new rows with the sampled values under target_column_name.

    Raises:
        ValueError: If n is negative or larger than the source column size
                    when replace=False.
        KeyError: If source_column_name does not exist in source_df.
        TypeError: If source_df or target_df are not pandas DataFrames.
    """
    # --- Input Validation ---
    if not isinstance(source_df, pd.DataFrame):
        raise TypeError("source_df must be a pandas DataFrame")
    if not isinstance(target_df, pd.DataFrame):
        raise TypeError("target_df must be a pandas DataFrame")
    if not isinstance(source_column_name, str):
        raise TypeError("source_column_name must be a string")
    if not isinstance(target_column_name, str):
        raise TypeError("target_column_name must be a string")
    if not isinstance(n, int) or n < 0:
        raise ValueError("n must be a non-negative integer")
    if source_column_name not in source_df.columns:
        raise KeyError(f"Column '{source_column_name}' not found in source_df")

    if n == 0:
        print("Warning: n is 0, no samples taken. Returning original target_df.")
        # Return a copy to maintain consistency of returning a new DataFrame
        return target_df.copy()

    # Check if sampling is possible (n vs source size) only if replace=False
    if not replace and n > len(source_df):
         raise ValueError(f"Cannot sample n={n} items without replacement "
                          f"from a column of size {len(source_df)}")

    # --- Core Logic ---
    try:
        # 1. Sample n values from the specified source column
        sampled_values = source_df[source_column_name].sample(n=n, replace=replace)

        # 2. Create a temporary DataFrame from the sampled values
        #    The column name in this temp DataFrame should be the target column name
        temp_df = pd.DataFrame({target_column_name: sampled_values})

        # 3. Append the temporary DataFrame rows to the target DataFrame
        #    ignore_index=True creates a new clean index for the combined DataFrame
        result_df = pd.concat([target_df, temp_df], ignore_index=True, sort=False)

        return result_df

    except ValueError as e:
         # Catch potential errors from .sample() if checks missed something
         print(f"Error during sampling: {e}")
         raise # Re-raise the error after printing
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        raise # Re-raise

iise_ap_queries = sample_and_append_rows(iise_df, 
                                      'abstract', 
                                      target_df=iise_queries, 
                                      target_column_name='Prompt',
                                      n=20,
                                      replace=False)

informs_ap_queries = sample_and_append_rows(informs_df, 
                                      'abstract', 
                                      target_df=informs_queries, 
                                      target_column_name='Prompt',
                                      n=20,
                                      replace=False)

display(iise_ap_queries.tail(25))
display(informs_ap_queries.tail(25))
```
